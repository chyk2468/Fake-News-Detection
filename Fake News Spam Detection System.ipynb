```python
import re
```

```python
def wordopt(text):
  text=text.lower()
  text=re.sub(r'\[.*?\]','',text)
  text=re.sub(r'https?-://\S+|www\.\S+\n\d\w','',text)
  return text
```

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "True News"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorizer.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_DTC = DTC.predict(new_xv_test)
  pred_RFC = rfc.predict(new_xv_test)
  pred_GBC = gbc.predict(new_xv_test)
  return print("\n\nLR Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {} \nGBC Prediction: {}".format(
      output_lable(pred_LR[0]),output_lable(pred_DTC[0]),output_lable(pred_RFC[0]),output_lable(pred_GBC[0])))
```
